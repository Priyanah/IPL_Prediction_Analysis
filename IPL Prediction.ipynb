# **Data Cleaning (removing unwanted columns)**

```python
clms_remove = ['mid', 'batsman', 'bowler', 'striker', 'non-striker','venue']
df.drop(labels=clms_remove, axis=1,inplace=True)
```

```python
df.head()
```

```python
df['bat_team'].unique()
```

```python
#keeping only consistent teams.
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Punjab Kings','Royal Challengers Bangluru',
                    'Delhi Capitals', 'Sunrisers Hyderabad']
```

```python
# only taking data of consistent teams of bat_team & bowl_team.
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
#removing the first 5 overs data in every match
df = df[df['overs']>=5.0]
df.head()
```

```python
print(df['bat_team'].unique())
print(df['bowl_team'].unique())
```

```python
# converting the column 'date' from string into datatime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%d-%m-%Y'))
```

```python
df.head()
```

```python
# converting categorical fetures using OneHotEncoding.
one_hot_encode = pd.get_dummies(data = df, columns = ['bat_team', 'bowl_team'])
```

```python
one_hot_encode.head()
```

```python
one_hot_encode.columns
```

```python
print(df.shape)
```

```python
rearrange_clms = one_hot_encode[["date", "bat_team_Chennai Super Kings", "bowl_team_Chennai Super Kings", "bat_team_Delhi Capitals", "bowl_team_Delhi Capitals", "bat_team_Kolkata Knight Riders", "bowl_team_Kolkata Knight Riders", "bat_team_Mumbai Indians", "bowl_team_Mumbai Indians",
                                 "bat_team_Punjab Kings", "bowl_team_Punjab Kings", "bat_team_Rajasthan Royals","bowl_team_Rajasthan Royals", "bat_team_Royal Challengers Bangluru", "bowl_team_Royal Challengers Bangluru", "bat_team_Sunrisers Hyderabad", "bowl_team_Sunrisers Hyderabad",
                                 "overs", "runs_last_5", "runs", "wickets_last_5", "wickets", "total"]]
```

```python
print(rearrange_clms)
```

```python
rearrange_clms.columns
```

```python
# X
X_train = rearrange_clms.drop(labels='total', axis = 1)[rearrange_clms['date'].dt.year <= 2016] # all data before & till 2016
X_test = rearrange_clms.drop(labels='total', axis = 1)[rearrange_clms['date'].dt.year >= 2016]  # all data after & till 2017
# y
y_train = rearrange_clms[rearrange_clms['date'].dt.year <= 2016]['total'].values
y_test = rearrange_clms[rearrange_clms['date'].dt.year >= 2017]['total'].values
```

```python
# now "date" clm is not required, it was only for train and test purpose.
X_train.drop(labels = 'date', axis = True, inplace = True)
X_test.drop(labels = 'date', axis = True, inplace = True)
```

```python
# we will be using RandomForest Regression
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
```

```python
#creating a pickle file for the classifier
filename = 'first-innings-score-rf-model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```
